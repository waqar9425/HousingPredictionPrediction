```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_Y,test_Y = train_test_split(X,Y,random_state=0,test_size=0.3)
```

```python
train_X.shape,train_Y.shape,test_X.shape,test_Y.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
from sklearn import metrics
rf.fit(train_X,train_Y)
pred=rf.predict(test_X)
print('accuracy score of rf mode is',metrics.r2_score(test_Y,pred))
print('accuracy score of rf mode is',metrics.mean_absolute_error(test_Y,pred))
print('accuracy score of rf mode is',metrics.r2_score(test_Y,pred))
```

```python
from sklearn.model_selection import cross_val_score
scoring_metrics = metrics.make_scorer(metrics.mean_squared_error, 
                              greater_is_better=False
                             )
score=cross_val_score(rf,X,Y,cv=10,scoring=scoring_metrics)
print(-score.mean())
```
